```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0, 10, 400)
y1 = x**2
y2 = x**3
y3 = np.sqrt(x)
plt.plot(x, y1, label='y = x²', color='blue', linestyle='-')
plt.plot(x, y2, label='y = x³', color='green', linestyle='--')
plt.plot(x, y3, label='y = √x', color='red', linestyle=':')
plt.title('Plot of y = x², y = x³, and y = √x')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
x = np.linspace(0, 2 * np.pi, 400)
y1 = np.sin(x)
y2 = np.cos(x)
fig, (x1, x2) = plt.subplots(2, 1, sharex=True, figsize=(8, 6))
x1.plot(x, y1, color='blue')
x1.set_title('y = sin(x)')
x2.plot(x, y2, color='green')
x2.set_title('y = cos(x)')
x2.set_xlabel('x')
fig.suptitle('Sine and Cosine Functions', fontsize=14)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
subjects = ['Math', 'Science', 'English']
scores = [85, 90, 78]
std_devs = [5, 7, 4]
plt.bar(subjects, scores, yerr=std_devs, capsize=10, color=['skyblue', 'lightgreen', 'salmon'])
plt.title('Average Scores with Standard Deviation')
plt.ylabel('Scores')
plt.ylim(0, 100)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
x = np.random.uniform(0, 10, 100)
y = np.random.uniform(0, 10, 100)
distance = np.sqrt(x**2 + y**2)
plt.scatter(x, y, c=distance, cmap='viridis')
plt.colorbar(label='Distance from Origin')
plt.title('Scatter Plot Colored by Distance from Origin')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

```python
data1 = np.random.normal(loc=0, scale=1, size=1000)
data2 = np.random.normal(loc=2, scale=1.5, size=1000)
plt.hist(data1, bins=30, alpha=0.6, label='Data1 (mean=0, std=1)', color='blue')
plt.hist(data2, bins=30, alpha=0.6, label='Data2 (mean=2, std=1.5)', color='orange')
plt.title('Histograms of Two Normal Distributions')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()
```

```python
x = np.linspace(0, 2 * np.pi, 400)
y_sin = np.sin(x)
y_cos = np.cos(x)
plt.plot(x, y_sin, label='sin(x)', color='blue', linestyle='-')
plt.plot(x, y_cos, label='cos(x)', color='red', linestyle='--')
plt.title('Sine and Cosine Waves')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
plt.grid(True)
plt.show()
```

```python
x = np.random.uniform(0, 10, 100)
y = np.random.uniform(0, 10, 100)
distance = np.sqrt(x**2 + y**2)
plt.scatter(x, y, c=distance, s=distance*10, cmap='plasma', alpha=0.7)
plt.colorbar(label='Distance from Origin')
plt.title('Scatter Plot with Size and Color Variations')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.grid(True)
plt.show()
```

```python
x = np.linspace(0, 10, 400)
y1 = x
y2 = x**2
y3 = x**3
fig, (ax1, ax2, ax3) = plt.subplots(3, 1,  sharex=True, figsize=(8, 10))
ax1.plot(x, y1, color='blue')
ax1.set_title('y = x')
ax1.grid(True)
ax2.plot(x, y2, color='green')
ax2.set_title('y = x²')
ax2.grid(True)
ax3.plot(x, y3, color='red')
ax3.set_title('y = x³')
ax3.set_xlabel('x')
ax3.grid(True)
plt.tight_layout()
plt.show()
```

```python
x = np.linspace(0, 5, 400)
y = -x**2 + 4*x + 1
x_max = 2 
y_max = -x_max**2 + 4*x_max + 1
plt.plot(x, y, label='y = -x² + 4x + 1')
plt.scatter(x_max, y_max, color='red')
plt.annotate(f'Maximum\n({x_max}, {y_max})', xy=(x_max, y_max), xytext=(x_max+0.5, y_max-5),
             arrowprops=dict(facecolor='black', shrink=0.05))
plt.title('Plot with Maximum Point Annotated')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
values = [5, 7, 3, 8, 6]
errors = [0.5, 0.7, 0.3, 0.2, 0.4]
bars = range(len(values))
plt.bar(bars, values, yerr=errors, capsize=5, color='teal')
plt.xlabel('Bar Number')
plt.ylabel('Height')
plt.title('Bar Plot with Error Bars')
plt.show()
```

```python
labels = ['Apple', 'Samsung', 'Others']
sizes = [40, 35, 25]
explode = [0.1, 0, 0]
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, explode=explode, shadow=True)
plt.title('Market Share')
plt.axis('equal')
plt.show()
```

```python
x = np.linspace(0, 4 * np.pi, 200)
y = np.sin(x)
tick_positions = np.arange(0, 4 * np.pi + 0.1, np.pi / 2)
tick_labels = [f'{i}π/2' for i in range(9)]
plt.plot(x, y)
plt.xticks(tick_positions, tick_labels)
plt.ylim(-1.5, 1.5)
plt.title('Customized Ticks and Axis Limits')
plt.show()
```

```python
labels = ['Group1', 'Group2', 'Group3', 'Group4']
men_means = [5, 7, 6, 8]
women_means = [6, 9, 5, 7]
plt.bar(labels, men_means, label='Men')
plt.bar(labels, women_means, bottom=men_means, label='Women')
plt.ylabel('Scores')
plt.title('Stacked Bar Plot')
plt.legend()
plt.show()
```
