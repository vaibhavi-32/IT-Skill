```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]
plt.plot(x, y)
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Simple Line Chart")
plt.show()
```

```python
import matplotlib.pyplot as plt
fruits = ['Apples', 'Bananas', 'Cherries']
uantity = [10, 15, 7]
plt.bar(fruits, uantity, color='skyblue')
plt.title("Fruit uantities")
plt.xlabel("Fruits")
plt.ylabel("uantity")
plt.show()
```

```python
import matplotlib.pyplot as plt
subjects = ['Math', 'Science', 'English', 'History']
sizes = [30, 25, 25, 20]
plt.pie(sizes, labels=subjects, autopct='%1.1f%%', startangle=140)
plt.title("Subject Preference")
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1, 2, 3]
y1 = [2, 4, 6]
y2 = [1, 2, 3]
plt.plot(x, y1, label='Line 1')
plt.plot(x, y2, label='Line 2')
plt.legend()
plt.title("Two Lines with Legend")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
import matplotlib.pyplot as plt
x = ['A', 'B', 'C']
y = [5, 7, 3]
plt.bar(x, y, color='green')
plt.title("Sample Bar Chart")
plt.savefig("chart.png") # Saves the chart to current directory
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4]
y = [2, 3, 5, 7]
plt.plot(x, y, 'r--') # red dashed line
plt.title("Red Dashed Line")
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]
y2 = [2, 3, 4, 5]
plt.subplot(1, 2, 1) # 1 row, 2 columns, 1st plot
plt.plot(x, y1)
plt.title("Line Chart")
plt.subplot(1, 2, 2) # 1 row, 2 columns, 2nd plot
plt.bar(x, y2)
plt.title("Bar Chart")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
data = [1, 2, 2, 3, 3, 3, 4, 4, 5]
plt.hist(data, bins=5, color='orange', edgecolor='black')
plt.title("Histogram")
plt.xlabel("Value")
plt.ylabel("Freuency")
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
plt.plot(x, y)
plt.annotate("This is 3,9", xy=(3, 9), xytext=(3, 15),
arrowprops=dict(facecolor='black', shrink=0.05))
plt.title("Annotated Plot")
plt.show()
```
