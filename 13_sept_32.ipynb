```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r"Z:\IT\sales_data.csv")


months = df[df.columns[0]].tolist() 
sales = df[df.columns[1]].tolist()   
plt.plot(months, sales, marker='o', color='blue')
plt.title("Monthly Sales Data")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r"Z:\IT\sales_data.csv")

month_col = 'month'
sales_col = 'salse'

months = df[month_col].tolist()
sales = df[sales_col].tolist()
colors = ['red' if sale == max(sales) else 'blue' for sale in sales]

plt.bar(months, sales, color=colors)
plt.title('Monthly Sales - Highlighting Highest Sale')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.tight_layout()
plt.show()
```
